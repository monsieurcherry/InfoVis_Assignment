```python
# Load dataset.
df = pd.read_csv("https://raw.githubusercontent.com/monsieurcherry/InfoVis_Assignment/main/food_consumption.csv")

# Perform aggregation and calculation.
aggregated_df = df.groupby("country").agg(
    total_consumption=("consumption", "sum"),
    total_emission=("co2_emmission", "sum")
).reset_index()

aggregated_df["total_consumption_emission"] = aggregated_df["total_consumption"] + aggregated_df["total_emission"]

display(aggregated_df.head())
```

```python
import requests

# Load the GeoJSON data.
geojson_url = "https://raw.githubusercontent.com/monsieurcherry/InfoVis_Assignment/main/world_map.json"
response = requests.get(geojson_url)
geojson_data = response.json()

# Extract country names from GeoJSON data.
geojson_countries = set(feature["properties"]["name"] for feature in geojson_data["features"])

# Extract country names from the CSV data.
csv_countries = set(aggregated_df["country"])

# Find missing countries.
missing_countries = csv_countries - geojson_countries

print("Countries missing in GeoJSON from CSV:")
print(missing_countries)
```

```python
# List of countries to print.
countries_to_print = [
    'Mauritius', 'French Polynesia', 'Maldives', 'Madagascar', 'New Zealand', 'India'
]

# Print rows for the specified countries.
selected_rows = aggregated_df[aggregated_df['country'].isin(countries_to_print)]
display(selected_rows)
```

```python
# Group specified countries with their corresponding groups.
group_mapping = {
    'Mauritius': 'Madagascar',
    'French Polynesia': 'New Zealand',
    'Maldives': 'India'
}

# Rename specified countries.
rename_mapping = {
    'USA': 'United States of America',
    'Swaziland': 'eSwatini',
    'Bosnia and Herzegovina': 'Bosnia and Herz.',
    'Hong Kong SAR. China': 'Hong Kong',
    'Taiwan. ROC': 'Taiwan',
    'Czech Republic': 'Czechia',
    'Macedonia': 'North Macedonia'
}

# Apply group and rename mappings.
aggregated_df['country'] = aggregated_df['country'].replace(group_mapping).replace(rename_mapping)

# Perform aggregation on the specified columns.
aggregated_df = aggregated_df.groupby("country").agg(
    total_consumption=("total_consumption", "sum"),
    total_emission=("total_emission", "sum"),
    total_consumption_emission=("total_consumption_emission", "sum")
).reset_index()

# Display the updated DataFrame.
display(aggregated_df)

# List of countries to print.
countries_to_print_processed = [
    'Madagascar', 'New Zealand', 'India', 'United States of America', 'eSwatini', 'Bosnia and Herz.', 'Hong Kong', 'Taiwan', 'Czechia', 'North Macedonia'
]

# Print rows for the specified countries.
selected_rows_processed = aggregated_df[aggregated_df['country'].isin(countries_to_print_processed)]
display(selected_rows_processed)

# Extract country names from the processed CSV data.
processed_countries = set(aggregated_df["country"])

# Find missing countries.
missing_countries = processed_countries - geojson_countries

print("Countries missing in GeoJSON from CSV:")
print(missing_countries)
```

```python
# Save the processed data to a new CSV.
aggregated_df.to_csv("processed_food_consumption.csv", index=False)
```
